```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
# conda install -c conda-forge mlxtend
```

```python
book = pd.read_csv('C://Users//Admin//Desktop//Data//Assignments New//Association Rules//book.csv')
book
```

```python
#preprocessing

df=pd.get_dummies(book)
df
```

```python
df.describe()
```

```python
#Apriori Algorithm

frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets1 = apriori(df, min_support=0.3, use_colnames=True)
frequent_itemsets1
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5)
rules
```

```python
rules1 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules1
```
